```python
saved_output_dir  = 'C:/SimplonIA/ProjetBrasil/documents/repertoire.images'

def check_for_directory(directory, create_if_needed=True):
    """Voir si le répertoire existe. S'il n'existe pas il est créé."""
    path_exists = os.path.exists(directory)
    if path_exists:
        if not os.path.isdir(directory):
            raise Exception("Trouvé le nom "+directory+" mais c'est un fichier, pas un répertoire")
            return False
        return True
    if create_if_needed:
        os.makedirs(directory)
    return path_exists

check_for_directory(saved_output_dir)
```

```python
def render_mpl_table( data, 
                      col_width=3.0, 
                      row_height=0.625, 
                      font_size=14, 
                      header_color='#40466e', 
                      row_colors=['#f1f1f2', 'w'], 
                      edge_color='w', 
                      bbox=[0, 0, 1, 1], 
                      header_columns=0, 
                      ndec = 2,
                      fichierPdf = None,
                      titre='',
                      ax=None, **kwargs): 
    if ax is None: 
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height]) 
        fig, ax = plt.subplots(figsize=size) 
        ax.axis('off') 
    
    plt.title(titre, fontsize = 32);
    
    mpl_table = ax.table(cellText=data.values.round(ndec), bbox=bbox, colLabels=data.columns, **kwargs) 

    mpl_table.auto_set_font_size(False) 
    mpl_table.set_fontsize(font_size) 

    for k, cell in six.iteritems(mpl_table._cells): 
        cell.set_edgecolor(edge_color) 
        if k[0] == 0 or k[1] < header_columns: 
            cell.set_text_props(weight='bold', color='w') 
            cell.set_facecolor(header_color) 
        else: 
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ]) 
    if fichierPdf is not None :
        fichierPdf.savefig()       
    return ax 
```

# La variation en temps des évaluations et du CAB par client

<img src="https://raw.githubusercontent.com/rbizoi/Ecole-IA-Strasbourg/master/images/documents/e-brasil02.png" width="800">

```python
query = """
SELECT
    order_id,
    c.customer_unique_id client,
    o.order_status,
    extract(year from o.order_purchase_timestamp) as annee,
    extract(month from o.order_purchase_timestamp) as mois,
    extract(year from o.order_purchase_timestamp)*100 + extract(month from o.order_purchase_timestamp) as annee_mois,
    p.value_boleto+p.value_credit_card+p.value_debit_card+p.value_voucher as ca,
    r.review_1 + r.review_2*2 + r.review_3*3 + r.review_4*4 + r.review_5*5  NoteTotale,
    c.customer_zip_code_prefix cp,
    g.geolocation_city ville,
    g.nom_etat etat,
    g.lat_mean,
    g.lng_mean
FROM
    order_payments  p 
         JOIN orders o using(order_id)
         JOIN order_reviews r using(order_id)
         JOIN customers c using(customer_id)
         JOIN geolocation g on (c.customer_zip_code_prefix = g.geolocation_zip_code_prefix)
"""
donnees = pd.read_sql_query(query, connection)
donnees.shape
```

```python
d_jours = {1:"lundi",2:"mardi",3:"mercredi",4:"jeudi",5:"vendredi",6:"samedi",7:"dimanche"}
d_mois  = {1:"janvier",2:"février",3:"mars",4:"avril",5:"mai",6:"juin",7:"juillet",8:"août",9:"septembre",10:"octobre",11:"novembre",12:"décembre"}
listeJours = ["lundi","mardi","mercredi","jeudi","vendredi","samedi","dimanche"]
listeMois = ["janvier","février","mars","avril","mai","juin","juillet","août","septembre","octobre","novembre","décembre"]
```

```python
customer_id = {cli:i for i,cli in enumerate(donnees.sort_values('client').client.unique())}
donnees['client'] = donnees.client.apply(lambda x: customer_id[x])
```

# Las ventes 2018  

```python
affichage = donnees[donnees.annee == 2018][['order_id','lat_mean','lng_mean']]

latitude  = (affichage.lat_mean.max() + affichage.lat_mean.min())/2
longitude = (affichage.lng_mean.max() + affichage.lng_mean.min())/2
```

```python
map_brasil = folium.Map(location=[latitude,longitude],zoom_start=4.3)#[48.85, 2.34] Paris
heat_data = [[row['lat_mean'],row['lng_mean']] for index, row in affichage.iterrows()]
folium.plugins.HeatMap(heat_data).add_to(map_brasil)

map_brasil.save("Ventes.Regions.html")
map_brasil
```

```python
def strip_accents_unicode(s):
    normalized = unicodedata.normalize('NFKD', s)
    return ''.join([c for c in normalized if not unicodedata.combining(c)])
```

```python
affichage = donnees.groupby('etat').agg({'order_id':'count','ca':'sum','notetotale':'mean',
                                         'lat_mean':'mean', 'lng_mean':'mean'})
```

```python
affichage.columns = ['commandes','ca','note_moyenne','latitude','longitude']
affichage.reset_index(inplace=True)
```

```python
affichage.etat = affichage.etat.map(lambda x: strip_accents_unicode(x))
```

```python
affichage.columns
```

```python
map_ebrasil = folium.Map(location=[latitude ,longitude],zoom_start=3)

for val in affichage.itertuples():
    map_ebrasil.add_child(  folium.CircleMarker(location=[val.latitude,val.longitude], 
          tooltip="<p><i><b>%s</b></i></p><p>Nb_Commandes&nbsp;:&nbsp;%d</p><p>Somme_CA&nbsp;:&nbsp;%d</p><p>NoteMoyenne&nbsp;:&nbsp;%d</p>"%(
                                               val.etat, val.commandes, val.ca, 
                                               val.note_moyenne),
                        #color='Blue',
                        weight = 1,
                        #stroke=False,
                        fill_color='#132b5e',
                        fill_opacity=0.3,                      
                        radius=val.commandes/500))

map_ebrasil.save("Informations Régions--"+
                 datetime.datetime.today().strftime('%Y-%m-%d')
                 +".html")
map_ebrasil
```
